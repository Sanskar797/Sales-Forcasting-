```python
data=pd.DataFrame(list(df.items()),columns=['customer','Items'])
```

```python
data.head()
```

```python
data['customer']=data['customer'].apply(lambda x:x.lstrip('customer'))
```

```python
Items_List=[]
for i in data['Items']:
    item=[]
    for j in i:
        rec=j.lstrip('Item_')
        item.append(rec)
    Items_List.append(item)
                
#data['Items']=data['Items'].apply([lambda x:x.lstrip('Item_') for x in data['Items']])
```

```python
data['Items']=Items_List
```

```python
data.head()
```

```python
data = pd.melt(data.set_index('customer')['Items'].apply(pd.Series).reset_index(), 
             id_vars=['customer'],
             value_name='Items') \
    .dropna().drop(['variable'], axis=1) \
    .groupby(['customer', 'Items']) \
    .agg({'Items': 'count'}) \
    .rename(columns={'Items': 'purchase_count'}) \
    .reset_index() \
    .rename(columns={'Items': 'productId'})
data['productId'] = data['productId'].astype(np.int64)
```

```python
data.shape
data.head()
```

```python
ratings_utility_matrix = data.pivot_table(values='purchase_count', index='customer', columns='productId', fill_value=0)
ratings_utility_matrix
```

```python
ratings_utility_matrix.shape
```

```python
X = ratings_utility_matrix.T
X
```

```python
X1 = X.copy()
```

```python
SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(X)
decomposed_matrix.shape
```

```python
correlation_matrix = np.corrcoef(decomposed_matrix)
correlation_matrix.shape
```
